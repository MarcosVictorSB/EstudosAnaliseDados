```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
sns.set(style='darkgrid')
```

```python
dataset = pd.read_csv('Segundo_semestre_2017.csv', sep=';', encoding='latin1')
```

```python
dataset.head()
```

```python
dataset.count() # contagem dos registros por atributos existente no dataset
```

```python
# Quantidade de reclamações por sexo
dataset['Sexo'].value_counts()
```

```python
plt.figure(figsize=(12,5))
sns.countplot(x='Sexo', data=dataset)
```

```python
# Qual a faixa etária média dos consumidores
dataset.groupby('Sexo')['Faixa Etária'].value_counts()
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(x='Faixa Etária', hue='Sexo', data=dataset,)
```

```python
# Qual grupo de problema mais comum
dataset['Grupo Problema'].value_counts()
```

```python
plt.figure(figsize=(16, 8))
sns.countplot(x='Grupo Problema', data=dataset, order=dataset['Grupo Problema'].value_counts().index)
```

```python
# Qual o segmento de mercado mais afetado
segmento = dataset[dataset['Grupo Problema'] == 'Cobrança / Contestação']
segmento['Segmento de Mercado'].value_counts()
```

```python
plt.figure(figsize=(20,14))
sns.countplot(y=segmento['Segmento de Mercado'], data=segmento)
```

```python
# Existe alguma correlação entre o tempo de resposta e a nota do consumidor
# dataset = dataset.drop('Total', axis=1)
corr = dataset.corr()
sns.heatmap(corr, cmap='YlGnBu', annot=True)
```
